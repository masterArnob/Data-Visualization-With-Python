# Area Plots, Histograms, and Bar Charts

## Objectives

*   Create and customize
    - Area plot
    - Histogram
    - Bar charts on a dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Canada.csv')
df.head()
```

```python
df.set_index('Country', inplace=True)
df.head()
```

```python
years = list(map(str, range(1980, 2014)))
years
```

# **Area Plot**

```python
df.sort_values(['Total'], ascending=False, axis=0, inplace=True)
df.head()
```

```python
df_top5 = df.head()
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python

df_top5.plot(kind='area', stacked=False, figsize=(20, 10), alpha=0.25)
plt.xlabel('Years')
plt.ylabel('Immigration Trend Of Top 5 Countries')
plt.show()
```

**Question: Use the scripting layer to create a stacked area plot of the 5 countries that contributed the least to immigration to Canada from 1980 to 2013. Use a transparency value of 0.45.**

```python
years = list(map(str, range(1980, 2014)))
years
df_bottom = df.tail(5)
df_bottom = df_bottom[years].transpose()
df_bottom.plot(kind='area', figsize=(20, 10), alpha=0.45)
plt.xlabel('Years')
plt.ylabel('Immigration')
plt.title('5 countries that contributed the least to immigration')
plt.show()
```

**Question: Use the artist layer to create an unstacked area plot of the 5 countries that contributed the least to immigration to Canada from 1980 to 2013. Use a transparency value of 0.55.**

```python
df_bottom = df.tail(5)
years = list(map(str, range(1980, 2013)))
df_bottom = df_bottom[years].transpose()
df_bottom.plot(kind='area', figsize=(20, 10), alpha=0.55)
plt.xlabel('Years')
plt.ylabel('Immigration')
plt.title('5 countries that contributed the least to immigration to Canada')
plt.show()
```

# **Histogram**

**Question: What is the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?**

```python
df_hist = df['2013']
```

```python
df_hist.plot(kind='hist', figsize=(20, 10))
plt.xlabel('Number Of Immigrants')
plt.ylabel('Number Of Countries')
plt.title(' frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013')
plt.show()
```

# **Bar Chart**
**Question: Let's compare the number of Icelandic immigrants (country = 'Iceland') to Canada from year 1980 to 2013.**

```python
df = pd.read_csv('/content/Canada.csv')
df.set_index('Country', inplace=True)
df.head()
years = list(map(str, range(1980, 2013)))
df_iceland = df.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(20, 10))
plt.xlabel('Years')
plt.ylabel('Number Of Immigrants')
plt.title('Icelandic immigrants (country = \'Iceland\') to Canada from year 1980 to 2013')
plt.show()
```
